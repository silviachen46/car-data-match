```python
def ShortDelInd(file):
    List=[]#时间戳的容器
    Index=[]#待删除的重复数据
    for i in pd.Series(file.iloc[:,0]):#对时间戳遍历 
        if i in List:
            Index.append(file[file.time==i].index.tolist()[0])
            continue
        List.append(i)
    return Index#得到待删除序列 去重的序列
def ShortDelLis(file):
    List=[]#时间戳的容器
    Index=[]#待删除的重复数据
    for i in pd.Series(file.iloc[:,0]):#对时间戳遍历 
        if i in List:
            Index.append(file[file.time==i].index.tolist()[0])
            continue
        List.append(i)
    return List#得到待删除序列 去重的序列
```

```python
def DropInd(Ind,file):
    file=file.drop(index=Ind)
    file.reset_index(drop=True,inplace=True)
    return file
```
